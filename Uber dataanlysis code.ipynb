# Data Preprocessing

```python
# df['PURPOSE'].fillna('Not',inplace=True)
df['PURPOSE'] = df['PURPOSE'].ffill()
```

# Changing Dataformat

```python
df['START_DATE'] = pd.to_datetime(df['START_DATE'],errors='coerce')
df['START_DATE'] = df['START_DATE'].ffill()

```

```python
df['END_DATE'] = pd.to_datetime(df['END_DATE'],errors='coerce')
df['END_DATE'] = df['END_DATE'].ffill()

```

```python
df.info()
```

```python
df
```

```python
from datetime import datetime
df['Date'] = pd.DatetimeIndex(df['START_DATE']).date
```

```python
df['Time'] = pd.DatetimeIndex(df['START_DATE']).hour
```

```python
df.info()
```

```python
df.head()
```

```python
df['Day-Night'] = pd.cut(x=df['Time'],bins=[0,10,13,15,24],labels=['Morning','Afternoon','Evening','Night'])
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.isna().sum()
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,5))

# First subplot
plt.subplot(1,2,1)
sns.countplot(x='CATEGORY', data=df)
plt.title("Category Count")

# Second subplot
plt.subplot(1,2,2)
sns.countplot(y='PURPOSE', data=df)
plt.title("Purpose Count")

plt.show()
```

```python
sns.countplot(df['Day-Night'])
plt.show()
```

```python
df['Month'] = pd.DatetimeIndex(data=df['START_DATE']).month
Month_label = {1.0:'Jan',2.0:'Feb',3.0:'Mar',4.0:'Apr',
               5.0:'May',6.0:'June',7.0:'July',8.0:'Aug',
               9.0:'Sep',10.0:'Oct',11.0:'Nov',12.0:'Dec'}

df['Month'] = df.Month.map(Month_label)
count = df.Month.value_counts(sort=False)
```

```python
df.head()
```

```python
dtf = pd.DataFrame({'Month': count.values,
                   'Value Count': df.groupby('Month',sort=False)['MILES'].max()})
p = sns.lineplot(dtf)
p.set(xlabel='Month',ylabel='Value Count')
```

```python
df['Day'] = pd.DatetimeIndex(df['START_DATE']).weekday
Day_label = {0:'Monday',1:'Tuesday',2:'Wedesday',3:'Thurday',
               4:'Friday',5:'Saturday',6:'Sunday'}

df['Day'] = df.Day.map(Day_label)
count = df.Day.value_counts()
```

```python
df.head()
```

```python
sns.barplot(x=count.index,y=count)

plt.xlabel=('Day')
plt.ylabel=('Value Count')
```
